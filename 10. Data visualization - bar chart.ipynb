# 10. Data visualization - bar chart 






To present our findings better we need to learn how to **visualize data**. 
**Bar chart** is a good way to do this.

## Step 1: Import the libraries 

Just like last time, we need to `import` libraries. Today we will use `pandas` again and a new library called `matplotlib`. Because of their long names we just call them `pd` and `plt` to make it easier. 

````python
import pandas as pd
import matplotlib.pyplot as plt

survey = pd.read_csv('/Users/nadzi/Documents/Datasets/Shanghai Thomas Survey/Shanghai_Thomas_Survey_datasets/CSV ready/survey.csv')
````

````python
survey
````

Last time, we learned how to calculate **percentages**.

````python
survey['Q15_text'].value_counts(normalize=True) * 100
````

To re-use them later the best thing to do is make a **variable**.

Variables are great! They are like boxes that can keep something inside.

````python
# Create a variable to calculate percentages 
Q15_percentages = survey['Q15_text'].value_counts(normalize=True) * 100
````

Now, we just need to use the variable name!

````python
Q15_percentages
````

## Making the bar chart

Use the code below to make the bar chart!

````python
# Create a bar chart
Q15_percentages.plot(kind='bar')
plt.xlabel('Response')
plt.ylabel('Percentage')
plt.title('Percentage Distribution of Q15_text Responses')
plt.show()
````

````python
# Create a list of colors (ensure the length matches the number of bars)
colors = ['red','green', 'blue', 'orange', 'purple']
Q15_percentages.plot(kind='bar', color=colors)
plt.xlabel('Response')
plt.ylabel('Percentage')
plt.title('Percentage of students who consider themselves good students')
plt.show()
````
