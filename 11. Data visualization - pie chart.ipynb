One way to show the results is by using **bar chart** like we did last time.




Today, we will make **pie charts**

To do this, we need to first calculate the percentages.

````python
# Create a variable to calculate percentages 
Q15_percentages = survey['Q15_text'].value_counts(normalize=True) * 100
````

After that, the easiest way to display a pie chart is:

````python

plt.pie(Q15_percentages)
plt.show()
````

But what if we want to know what each "slice" is about?

````python
#we add labels
plt.pie(Q15_percentages, labels=Q15_percentages.index)
plt.show()
````

But what if we want to know the % for each slice?

````python
# We add percentages!


plt.pie(Q15_percentages, labels=Q15_percentages.index, autopct='%1.1f%%')
plt.show()
````

But what if we want to change colors and add a title? 

````python

# Example custom colors (ensure the number of colors is the same as the number of answers)
colors = ['pink','magenta', 'violet', 'plum', 'purple']

plt.pie(Q15_percentages, labels=Q15_percentages.index, colors=colors, autopct='%1.1f%%')
plt.title("Question 15: I consider myself a good student.")
plt.show()
````
