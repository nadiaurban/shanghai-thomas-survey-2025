## Looking at columns

Looking at the table doesn't tell us much. If the table is too big - like our survey is - then we can't even see the whole table at once.

But we can learn from individual columns!

To see a column you need to use square brackets: `[]` and put the `column name` inside of it. For example:

`survey['column1']`

```python
survey['Q2_text']
```

Again, we only see a part of the table. 

What we want to know is how all the students answered!

To do that we need to count their answers. We  do this using `.value_counts()` 

```python
survey['Q2_text'].value_counts()
```

We can add `.value_counts()` to any column. 

```python
survey['Q3_text'].value_counts()
```

There's an even better way to make conclusions from our data. 

That is using procentages!

Saying '50% of the students like chocolate ice cream' is much better than saying '32 students like chocolate ice cream'. Using procentages we can give more general opinions.

```python
survey['Q7_text'].value_counts(normalize=True) 
```

```python
survey['Q7_text'].value_counts(normalize=True) * 100 
```
