### Sometimes we want to see answers divided by different groups. 

We might want to see if there is a difference between how students answered in **different grades**.

Or  we might want to see if there is a difference between boys and girls.

To see that we need to add `groupby`

```python
# Calculate percentages for Q3 grouped by Q1_text (grade)
# What's the living situation in different grades? %

Q3_grouped_by_grade = (survey.groupby('Q1_text')['Q3_text'].value_counts(normalize=True).unstack() * 100).round(0)

Q3_grouped_by_grade
```

In `groupby` we first add the question by which we are grouping. 

For example we write `('Q1_text')` first to group by Grade. 

Then we write the question we want to check, such as the living situation: `['Q3_text']`.

 First we use `()`, then we use `[]`.

-----


Okay! Let's check which grade is the most confident in speaking English.

To do that we need to take a look at `'Q7_text'` and group it by `'Q1_text.'` 

```python
Q7_grouped_by_grade = (survey.groupby('Q1_text')['Q7_text'].value_counts(normalize=True).unstack() * 100).round(0)

Q7_grouped_by_grade
```
