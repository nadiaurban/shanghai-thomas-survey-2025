# 13. Grouping data - bar charts

## Step 1. Import the libraries, load the survey and check if it works.

## You must run this code!!!

```python
import pandas as pd
import matplotlib.pyplot as plt

survey = pd.read_csv('/Users/nadzi/Documents/Datasets/Shanghai Thomas Survey/Shanghai_Thomas_Survey_datasets/CSV ready/survey.csv')
```

```python
survey
```

## We now know how to group our questions to compare them. 

For example we can group by grade (using`'Q1_text'`) or by gender (`Q2_text'`)

```python
#Question 7: Do you feel confident speaking English?

Q7_grouped_by_gender = (survey.groupby('Q2_text')['Q7_text'].value_counts(normalize=True).unstack() * 100).round(0)

Q7_grouped_by_gender
```

```python
#Question 7: Do you feel confident speaking English?

Q7_grouped_by_grade = (survey.groupby('Q1_text')['Q7_text'].value_counts(normalize=True).unstack() * 100).round(0)

Q7_grouped_by_grade
```

## The best way to compare this is to make a bar chart. 

With a bar chart we can easily see the differences!

The easiest way is just this short code:


```python
Q7_grouped_by_gender.T.plot(kind='bar')
plt.show()
```

## Of course we can make our chart much better!

We can **add colors**, **add X label** and a **Y label** as well as a **title**!

```python
# Plot
colors = ['pink','blue']

Q7_grouped_by_gender.T.plot(kind='bar', color=colors)
plt.xlabel('Student Response')
plt.ylabel('Percentage')
plt.title('Confidence in Speaking English by gender')
plt.legend(title="Gender")
plt.show()
```
