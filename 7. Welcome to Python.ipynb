# Rules for Setting Variables in Python

When setting variables in Python, there are a few rules to follow:

### 1. **Variable names must start with a letter or an underscore**:
   - You cannot start a variable name with a number.
   
   **Correct**:
   - `age = 12`
   - `_score = 100`
   
   **Incorrect**:
   - `1age = 12   # This will give an error!`

### 2. **Variable names can only contain letters, numbers, and underscores**:
   - No spaces, punctuation marks, or special characters allowed.
   
   **Correct**:
   - `player_score = 50`
   
   **Incorrect**:
   - `pl@yer-score = 50   # This will give an error!`

### 3. **Python is case-sensitive**:
   - This means `Age` and `age` are two different variables.
   
   **Correct**:
   - `Age = 20`
   - `age = 30`
