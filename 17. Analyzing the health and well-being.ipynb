## Our report will use colors inspired by our school's logo!

```python
demographics_colors = [
    "#A61D24",  # Maroon (primary)
    "#D44D52",  # Lighter maroon
    "#0D2757",  # Navy (secondary)
    "#B0B0B0",  # Medium grey (adjusted from very light grey)
    "#9E9E9E"   # Slightly darker grey
]

english_proficiency_colors = [
    "#0D2757",  # Navy (primary)
    "#2E4C88",  # Lighter navy
    "#A61D24",  # Maroon (accent)
    "#B0B0B0",  # Medium grey
    "#D0D0D0"   # Slightly lighter than the previous grey for contrast
]

school_life_palette = [
    "#0D2757",  # Dark navy (primary)
    "#2E4C88",  # Mid-tone blue
    "#4A90E2",  # Medium blue
    "#A8D0E6"   # Light blue
]

free_time_palette = [
    "#005f99",  # Deep, vibrant blue
    "#0073e6",  # Bright blue
    "#0099ff",  # Vivid blue
    "#00ccff"   # Clear, cool blue
]

def show_palette(colors, title="Palette"):
    """Display a horizontal series of colored bars to visualize a palette."""
    fig, ax = plt.subplots(figsize=(len(colors) * 1.5, 1.5))
    for i, color in enumerate(colors):
        ax.barh(0, width=1, left=i, color=color)
    ax.set_xlim(0, len(colors))
    ax.set_ylim(-0.5, 0.5)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(title, fontsize=14)
    plt.show()

# Visualize the updated palettes
show_palette(demographics_colors, "Demographics colors")
show_palette(english_proficiency_colors, "English Proficiency colors")
show_palette(school_life_palette, "School Life colors")
show_palette(free_time_palette, "Free Time colors")
```

# Part 1: Make simple bar charts and pie charts

## Bar chart code

```python
# Bar Chart
Q27_percentages = survey['Q27_text'].value_counts(normalize=True) * 100
Q27_percentages.sort_index().plot(kind='bar', color=school_life_palette)
plt.xlabel('Responses')
plt.ylabel('Percentage')
plt.title('Level of stress')
plt.show()
```

## Pie chart code

```python
# (Pie Chart)
Q22_percentages = survey['Q22_text'].value_counts(normalize=True) * 100
plt.pie(Q22_percentages, labels=Q22_percentages.index, colors=school_life_palette, autopct='%1.1f%%')
plt.title("How many hours do you  sleep at night?")
plt.show()

```
