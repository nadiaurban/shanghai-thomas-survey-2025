# Data Types in Python: String, Integer, and Float

In this notebook, we will learn about three common data types in Python:
- **String**: Represents text. We need to add quotations. ' or "
- **Integer**: Represents whole numbers like 10 or 5
- **Float**: Represents decimal numbers like 1.2 or 6.66


box.png

````python
# String variable: represents text data
greeting = "Hello, students!"

# Integer variable: represents whole numbers
year = 2025

# Float variable: represents decimal numbers
pi = 3.14159
````


- **String (`greeting`)**: The variable `greeting` holds a sequence of characters (text). In Python, text data is enclosed in quotes.
- **Integer (`year`)**: The variable `year` holds a whole number.
- **Float (`pi`)**: The variable `pi` holds a decimal number.




# Using `print()` in Python

The `print()` function in Python is used to display information. You can use it in two main ways:

- **Printing Directly:**  
  You can pass a string or other data type directly into the `print()` function, and it will display the value. For example:  
  ```python
  print("Hello, students!")
  print(2025)
  print(3.14159)


  - **Printing a variable:**  
  You can pass a variable into the `print()` function. In this case we do not need to use quotation " to display strings. For example:  
  ```python
  print(greeting)
  print(year)
  print(pi)
