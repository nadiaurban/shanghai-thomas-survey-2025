## Our report will use colors inspired by our school's logo!

```python
demographics_colors = [
    "#A61D24",  # Maroon (primary)
    "#D44D52",  # Lighter maroon
    "#0D2757",  # Navy (secondary)
    "#B0B0B0",  # Medium grey (adjusted from very light grey)
    "#9E9E9E"   # Slightly darker grey
]

english_proficiency_colors = [
    "#0D2757",  # Navy (primary)
    "#2E4C88",  # Lighter navy
    "#A61D24",  # Maroon (accent)
    "#B0B0B0",  # Medium grey
    "#D0D0D0"   # Slightly lighter than the previous grey for contrast
]

school_life_palette = [
    "#0D2757",  # Dark navy (primary)
    "#2E4C88",  # Mid-tone blue
    "#4A90E2",  # Medium blue
    "#A8D0E6"   # Light blue
]

free_time_palette = [
    "#005f99",  # Deep, vibrant blue
    "#0073e6",  # Bright blue
    "#0099ff",  # Vivid blue
    "#00ccff"   # Clear, cool blue
]

def show_palette(colors, title="Palette"):
    """Display a horizontal series of colored bars to visualize a palette."""
    fig, ax = plt.subplots(figsize=(len(colors) * 1.5, 1.5))
    for i, color in enumerate(colors):
        ax.barh(0, width=1, left=i, color=color)
    ax.set_xlim(0, len(colors))
    ax.set_ylim(-0.5, 0.5)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(title, fontsize=14)
    plt.show()

# Visualize the updated palettes
show_palette(demographics_colors, "Demographics colors")
show_palette(english_proficiency_colors, "English Proficiency colors")
show_palette(school_life_palette, "School Life colors")
show_palette(free_time_palette, "Free Time colors")
```

## Question 1: Grades

What is the number of students in each grade? 

```python
# Q1: Grade Distribution (Bar Chart)
Q1_counts = survey['Q1_text'].value_counts()
Q1_counts.sort_index().plot(kind='bar', color=demographics_colors)
plt.xlabel('Grade')
plt.ylabel('Number of students in each grade')
plt.title('Grade Distribution')
plt.show()
```

# Part 2: Grouping data 

## In this part we will group the answers by grades (`Q1_text`) and gender (`Q2_text`)



### Grouping Question 5 (number of sibilings) by different grades (question 1)

```python
# Grouping Q5 (Number of Siblings) by Q1 (Grade)
siblings_by_grade = (survey.groupby('Q1_text')['Q5_text'].value_counts(normalize=True).unstack() * 100).round(0)
siblings_by_grade.plot(kind='bar', color=demographics_colors)
plt.xlabel('Grade')
plt.ylabel('Percentage')
plt.title('Number of Siblings by Grade')
plt.legend(title="Number of Siblings")
plt.show()
```

### Grouping Question 3 (living situation) by gender

```python
# Group Q3 (Living Situation) by Q2 (Gender)
living_by_gender = (survey.groupby('Q2_text')['Q3_text']
                    .value_counts(normalize=True)
                    .unstack() * 100).round(1)

living_by_gender.plot(kind='bar', color=demographics_colors)
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.title('Living Situation by Gender')
plt.legend(title='Living Situation')
plt.show()
```
